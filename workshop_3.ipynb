## Parte 2: Polinomios de Lagrange

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:


### Ejercicio 1: Datos (0,0), (30,0.5), (60,√3/2), (90,1)


```python
x = np.linspace(0, 100, 400)

y = -6.054e-7 * x**3 - 1.9943e-5 * x**2 + 0.0178098 * x

x_p = [0, 30, 60, 90]
y_p = [0, 0.5, np.sqrt(3)/2, 1]

plt.plot(x, y)
plt.scatter(x_p, y_p, color='red')
plt.grid()
plt.legend()
plt.show()

```

$L_0(x)=\frac{(x-2)(x-3)}{2}$

$L_1(x)=-(x-1)(x-3)$

$L_2(x)=\frac{(x-1)(x-2)}{2}$


$P(x) = -\frac{1}{2}x^2 + \frac{5}{2}x - 1$
